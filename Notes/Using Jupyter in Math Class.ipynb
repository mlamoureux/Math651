# Using Jupyter in Math Class
### M. Lamoureux, March 11 2016

In the 2016 Winter term at University of Calgary, I started teaching a course on Industrial Mathematics (Math 651), with the goal of exposing students to some useful techniques for working on real industrial problems with a strong mathematical component.

One key step was to teach them to use a modern software tool called Jupyter, which allows one to combine text, graphical images, mathematical formulas, and functioning computer code all in one document. In fact, Jupyter allows one to create living documents where the users can play around with mathematics and document what they are doing, all in one single software package.

Jupyter works in the browser (Firefox, Chrome, Safari, etc) and uses a simple interface with the user. There is a choice of computational "kernels" that allow the user to write their code in standard languages such as R, Python, Julia, Matlab, and others. There is also a choice of infrastructure to host the system: beginners can use the site try.jupyter.org, more advanced users can use the PIMS hosting site jupyter.pims.math.ca or even install the system on their own computer (Windows, Mac OS, or Linux). 

### First impressions - instructor
I like the system a lot. In fact, I use it all the time now for creating lecture notes as well as computational examples. In Fall 2015 I created all my lecture notes for a complex analsyis course at UBC, which included a great deal of interesting numerical examples and complex mathematical graphics. Jupyter was great for this, in particular in that it seamlessly integrates English text, mathematical formulas, and computations with graphical output.

I especially like that I can easily write code in Matlab (which I have used for 20+ years), Python and Julia (which I am only just learning). By installing Jupyter on my own computer, I can also access audio devices on my computer, which is very useful for signal processing demonstations -- one of the things I do a lot in my work. (Note: this does not seem to work on PIMS Jupyter Hub, or on try.jupyter.org). 

Some draw-backs. 
- To launch Juptyer on a Mac, you need to open a terminal and type in something on the command line. I have a hard time remembering typed commands, so eventually I wrote an AppleScript file that I just click on, which does the launch commands for me
- I seem to have a hard time keeping track of my files. On try.jupyter.org, the files are somewhere out on the cloud and you have to upload and download your work. On jupyter.pims.math.ca, the files are on some server at UBC and it is not really obvious to me how to access them directly from my Mac. So I upload and download my work. On my own laptop's implementation of Jupyter, I usually can find the files, although Jupyter does not make the search easy.
- Jupyter notebook, and the computational kernels, can be really slow at times. If you have a big file with a lot of mathematical formulas, Jupyter will be slow each time you open the file. It is like it is re-compiling all your TeX code every time you open the file.
- the computational kernels (especially Julia) can be very slow at initializing. Say you want to use a new graphics package. You have to "Add" the package, which can mean waiting several minutes while the package is downloaded from the web, compiled, and installed. And God Forbid you should be disconnected from the Web. On my own laptop, eventually everything I use regulary gets installed and stored, so the waits go away. This is never the case on try.jupyter.ord (you get a fresh, un-packaged kernel each time), and on jupyter.pims.math.ca, the record is spotty -- sometime packages are saved and there for you next time, sometimes they are not.
- these "delays" can be embarrasing when you teach a class, as you will try something easy (plot sin(x)), and have to wait a few minutes while the system installs somethign it needs.
- there is currently no easy way to include a sketch in your notes. My work-around is to sketch on paper, take a picture with the camera on my laptop, and load that into the notebook. It works, but I would rather sketch with my mouse. (I think I will write code to do this!)

### Third impressions - students

A couple of students use Linux, and they told me it was trivial to install. Of the ten students, we see they are using Matlab, Python, and R as their programming languages. About eight of the ten are using their own installation of Jupyter, while two are using the PIMS site to host their work.



### Final thoughts

I am requiring the students to submit work on Jupyter notebooks, so they are actually being forced to use it. Their comfort zone is to use TeX for their formal reports, and MATLAB or R for their computations, and to cut-and-paste between the two if they need to add code or graphics. So this is a push on my part to get them to learn to use Jupyter effectively. In the long run, I think it is a better tool for communicating and collaborating on mathematics.

## Fixed point methods for solving ODEs

### M. Lamoureux March 4, 2016

This method is called Picard's method. The idea is to start with a first order equation (or system of equations) in the form
$$ y' = f(t,y(t)),  \mbox{ with } y(t_0) = y_o$$
and convert to an integral form
$$ y(t) = y_o + \int_{t_0}^t f(s,y(s)) \,ds. $$

The iteration runs by creating a sequence of functions $y_1(t), y_2(t), y_3(t), \ldots$ and hoping they converge to the answer. The first function in the sequence is
$$ y_1(t) = y_o \mbox{ that is, it is the constant function with value } y_o.$$
Then the next function $y_2$ is given by integrating $f(t,y_1(t))$, so
$$ y_2(t) = y_o + \int_{t_0}^t f(s,y_1(s)) \,ds, $$
and in gereral we will define $y_{k+1}$ in terms of the previous $y_k$ as
$$y_{k+1}(t) = y_o + \int_{t_0}^t f(s,y_k(s)) \,ds. $$
With suitable conditions on the function $f(t,y)$, we can show this sequence converges to a solution of the ODE.

Let's do this in code. We use Julia here, because I am trying to learn this language, which is apparently good for math. You could easily do this in Python or Matlab.


First, we add a plotting package, so we can plot our answers later. 

```julia
Pkg.add("PyPlot")
using PyPlot
```

Let's try to solve the simple ODE
$$ y' = 2y \mbox{ with } y(0) = 1.$$
We know the exact answer is the exponential function $y(t) = \exp(2t)$. 

For the numerical problem, we set up a time variable t (which is a vector of sample points from 0 to 1), a solutions variable y (which is an array, with one column for each approximate solution $y_k$, and a function formula $f = 2*y*dt$ which we will sum up in order to integrate.


```julia
t=linspace(0,1,100)    # the time variable, a vector of entries linearly spaced from 0 to 1.
dt = 1/100;            # step size for the integrals
y=zeros(100,10)        # an array for the solutions. Each column represents a function y_k

y[:,1] = 1             # this is our initial function guess, the constant function 1

for k = 1:9            
    f=2*y[:,k]*dt      # here we compute the integrand  2*y*dt
    f[1]=1             # this is y_o
    y[:,k+1] = cumsum(f,1)    # the cummulative sum is the Riemann sums of the integral
end

plot(t,y)              # now we plot the various functions that approximate the solution
ylim(.8,8)
```

It is easy to see in the functions above that the $y_1$ function is the constant 1 function, the $y_2$ function is the linear function going straight from 1 to 3, the next one is a quadratic, and so on. These functions quickly converge to something that looks like an exponential.

Let try a harder ODE, like this:
Let's try to solve the simple ODE
$$ y' = (1-t)y \mbox{ with } y(0) = 1.$$
We expect something like exponential growth for $t$ close to zero, but once $(1-t)$ becomes negative, we should get rapid decay. Let's see if the numerics show this. 


```julia
t0=0    # Let's set up our initial parameters with labels
t1=2
y0 = 1
npoints = 1000    # number of sample points along the time axis 
niters = 10       # number of iterations for Picard's method

t=linspace(t0,t1,npoints)    
dt = (t1-t0)/npoints;           
y=zeros(npoints,niters)        

y[:,1] = y0             # this is the initial function, the constant function y0

for k = 1:(niters-1)            
    f=(1-t).*y[:,k]*dt  # here we compute the integrand  (1-t)*y*dt
    f[1]=y0             # initial value in y0 + \int f(t,y)
    y[:,k+1] = cumsum(f,1)    # the cummulative sum is the Riemann sums of the integral
end

plot(t,y)              # now we plot the various functions that approximate the solution
```

Okay, we see in the above plots that the solution is growing up to t=1, then decaying. Looks good. 

But, you should try investigating longer intervals. This reveals that the behaviour can get weird for larger values of t. This suggests we have to be careful with Picard's method to be sure of convergence. 

Here in the code and plot below, we see what happens on the t-interval $[0,5]$.


```julia
t0=0
t1=5
y0 = 1
npoints = 10000
niters = 10

t=linspace(t0,t1,npoints)    
dt = (t1-t0)/npoints;           
y=zeros(npoints,niters)        

y[:,1] = y0            

for k = 1:(niters-1)            
    f=(1-t).*y[:,k]*dt  
    f[1]=y0             
    y[:,k+1] = cumsum(f,1)    
end

plot(t,y)              
ylim(-5,5)
```

We see in the above that the approximate solutions are starting to blow up. Why? 

It might be useful to solve the ODE exactly. It's pretty easy to find the exact solution is
$$y(t) = \exp(t-t^2/2),$$
which should decay rapidly to zero. 

Probably in the numerics, the approximate solution is going negative, which messes up everything.
